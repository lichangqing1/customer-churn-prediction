```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

from src.data_cleaning import prepare_clean_data
from src.feature_engineering import add_churn_features

PROJECT_PATH = Path.cwd()
FIGURE_PATH = PROJECT_PATH / "reports" / "figures"
FIGURE_PATH.mkdir(parents=True, exist_ok=True)
raw = pd.read_csv(PROJECT_PATH / "data" / "raw" / "WA_Fn-UseC_-Telco-Customer-Churn.csv")
master = add_churn_features(prepare_clean_data(raw))
master.head()
```

# 1. Target distribution

```python
overall_churn_rate = master["churn_flag"].mean()

churn_counts = master["churn"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(churn_counts.index, churn_counts.values)
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.savefig(f"{FIGURE_PATH}/churn_distribution.png", dpi=300)
plt.show()
```

# 2. Churn rate by contract

```python
contract_churn = master.groupby('contract').agg(
    total_customers = ("customer_id", "count"),
    churn_rate = ("churn_flag", "mean"),
    avg_monthly_charges = ("monthly_charges", "mean")
).reset_index()


plt.figure(figsize=(7, 4))
plt.bar(contract_churn['contract'], contract_churn['churn_rate'])
plt.title("Churn Rate by Contract Type")
plt.xlabel("Contract type")
plt.ylabel("Churn Rate")
plt.tight_layout()
plt.savefig(f"{FIGURE_PATH}/churn_rate_by_contract_type.png", dpi=300)
plt.show()
```

# 3. Churn by tenure group

```python
master['tenure_group'] = pd.cut(
    master['tenure'],
    bins=[0, 12, 24, 48, 72],
    labels=['0-12 months', '13-24 months', '25-48 months', '49-72 months'],
    include_lowest=True
    )

tenure_churn = master.groupby('tenure_group').agg(
    total_customers = ("customer_id", "count"),
    churn_rate = ("churn_flag", "mean"),
    avg_monthly_charges = ("monthly_charges", "mean")
).reset_index()

plt.figure(figsize=(7, 4))
plt.bar(tenure_churn['tenure_group'], tenure_churn['churn_rate'])
plt.title("Churn Rate by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Churn Rate")
plt.savefig(f"{FIGURE_PATH}/churn_rate_by_tenure_group.png", dpi=300)
plt.show()
```

# 4. Monthly charges by churn

```python
master['monthly_charge_group'] = pd.cut(
    master['monthly_charges'],
    bins=[0, 35, 70, 120],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True)

charge_churn = master.groupby('monthly_charge_group').agg(
    total_customers = ('customer_id', 'count'),
    churn_rate = ('churn_flag', 'mean')
).reset_index()

master.boxplot(column='monthly_charges', by='churn_flag', figsize=(7, 5))
plt.title('Monthly Charges by Churn')
plt.suptitle("")
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.tight_layout()
plt.savefig(f'{FIGURE_PATH}/monthly_charges_by_churn.png', dpi=300)
plt.show()
```

# 5. Churn rate by payment method

```python
payment_churn = master.groupby('payment_method').agg(
    total_customers = ('customer_id', 'count'),
    churn_rate = ('churn_flag', 'mean'),
    avg_monthly_charges = ('monthly_charges', 'mean')
).reset_index()

payment_plot = payment_churn.sort_values(by='churn_rate', ascending=False)

plt.figure(figsize=(10, 5))
plt.barh(payment_plot['payment_method'], payment_plot['churn_rate'])
plt.title('Churn Rate by Payment Method')
plt.xlabel('Churn Rate')
plt.ylabel('Payment Method')
plt.tight_layout()
plt.savefig(f'{FIGURE_PATH}/churn_rate_by_payment.png', dpi=300)
plt.show()
```
